```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
figures = ['figs/simulated_func_non-dynamic.png','figs/simulated_func_dynamic.png']
margin = 300
fig, axes = plt.subplots(2, 1, figsize=(18,6))
for figure, ax in zip(figures, axes.ravel()):
    img = mpimg.imread(f"{figure}")
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    ax.set(frame_on=False)
    
    ax.imshow(img)        
    #ax.set_aspect()    
plt.subplots_adjust(wspace=0, hspace=0)
#fig.tight_layout(rect=[-0.01, -0.02, 1.02, 1.02], w_pad=0.0, h_pad=0.0)
fig.tight_layout()
fig.savefig("figs/simulated_func.png", bbox_inches='tight',dpi=300)
```
