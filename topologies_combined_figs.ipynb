```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
figures = ['figs/sim_topo_func_comm_non_dyn.png','figs/sim_topo_func_comm_dyn.png']
margin = 10
fig, axes = plt.subplots(1, 2, figsize=(5,6))
for figure, ax in zip(figures, axes.ravel()):
    img = mpimg.imread(f"{figure}")
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    ax.set(frame_on=False)
    
    ax.imshow(img)        
    #ax.set_aspect(1)    

#fig.tight_layout(rect=[-0.01, -0.01, .01, .01], w_pad=0.0, h_pad=0.0)
plt.subplots_adjust(wspace=0, hspace=0)

fig.suptitle('Common input \n', fontsize=10)
fig.tight_layout(pad=0)
fig.savefig("figs/fig_topo_comm_func.png", bbox_inches='tight',dpi=300)
```
