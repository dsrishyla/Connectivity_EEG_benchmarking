```python
import glob
import xarray as xr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
methods = ['coh','ciplv','imcoh','wpli2']
```

```python
mean_lists_ec = {}
mean_lists_eo = {}
ec_means = {}
eo_means = {}
ec_stds = {}
eo_stds = {}
d_vals = {}
p_vals = {}

vn = ['fusiform-rh', 'fusiform-lh', 'lingual-rh', 'lingual-lh', 'cuneus-rh', 'cuneus-lh',
                      'lateraloccipital-rh', 'lateraloccipital-lh']

xarrays = glob.glob(f'/Users/lina_01/Desktop/mpi_lemon/undirected_outputs/coh/*_coh_EC.nc')
subjects = []
for array in xarrays:
    subject = array.split('/')[7].split('_')[0]
    subjects.append(subject)

fig, axes = plt.subplots(2,4,figsize=(15,5))

for method, ax in zip(methods, axes[0].ravel()):
    xarrays = glob.glob(f'/Users/lina_01/Desktop/mpi_lemon/undirected_outputs/{method}/*_{method}_EC.nc')
                
    xarrays_combined = []
    for array in xarrays:
        xarrays_combined.append(xr.open_dataarray(array))
    
    xarrays_combined = xr.concat(xarrays_combined, pd.Index(subjects, name="subjects"))
    xarrays_combined_vn = xarrays_combined.sel(region1 = vn, region2 = vn)

    mean_ec = []
    
    for subject in subjects:
        sub_data = xarrays_combined_vn.sel(subjects=subject)
        mean = sub_data.mean(dim=["bootstrap_samples","region1","region2","frequencies"]).values
        mean_ec.append(float(mean))

    mean_lists_ec[method] = mean_ec

    overall_mean_ec = np.mean(mean_ec)
    ec_means[method] = overall_mean_ec

    overall_std_ec = np.std(mean_ec)
    ec_stds[method] = overall_std_ec

    matrix_df = xarrays_combined.mean(dim=["bootstrap_samples","subjects","frequencies"]).values + xarrays_combined.mean(dim=["bootstrap_samples","subjects","frequencies"]).values.T
    sns.heatmap(matrix_df,ax = ax)
    ax.set_title(method)

for method, ax in zip(methods, axes[1].ravel()):
    xarrays = glob.glob(f'/Users/lina_01/Desktop/mpi_lemon/undirected_outputs/{method}/*_{method}_EO.nc')
                
    xarrays_combined = []
    for array in xarrays:
        xarrays_combined.append(xr.open_dataarray(array))
    
    xarrays_combined = xr.concat(xarrays_combined, pd.Index(subjects, name="subjects"))
    xarrays_combined_vn = xarrays_combined.sel(region1 = vn, region2 = vn)

    mean_eo = []
    
    for subject in subjects:
        sub_data = xarrays_combined_vn.sel(subjects=subject)
        mean = sub_data.mean(dim=["bootstrap_samples","region1","region2","frequencies"]).values
        mean_eo.append(float(mean))
    
    mean_lists_eo[method] = mean_eo
    
    overall_mean_eo = np.mean(mean_eo)
    eo_means[method] = overall_mean_eo
    
    overall_std_eo = np.std(mean_eo)
    eo_stds[method] = overall_std_eo

    matrix_df = xarrays_combined.mean(dim=["bootstrap_samples","subjects","frequencies"]).values + xarrays_combined.mean(dim=["bootstrap_samples","subjects","frequencies"]).values.T
    sns.heatmap(matrix_df,ax = ax)


for method in methods:

    wilcox, p = scipy.stats.wilcoxon(mean_lists_ec[method], mean_lists_eo[method])
    p_vals[method] = p
    
    pooled_sd = np.sqrt((((len(mean_lists_ec[method])-1)*(ec_stds[method]**2) + (len(mean_lists_eo[method])-1)*(eo_stds[method])**2))/(len(mean_lists_ec[method]) + len(mean_lists_eo[method]) - 2))
    cohens_d = (ec_means[method]- eo_means[method])/pooled_sd
    d_vals[method] = cohens_d
    

 
```

```python
p_vals
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,6))

ind = np.arange(len(ec_means))  # the x locations for the groups
width = 0.35  # the width of the bars

rects1 = ax.bar(ind - width/2, [ec_means['coh'], ec_means['ciplv'], ec_means['imcoh'],ec_means['wpli2']], width, yerr=[ec_stds['coh'],ec_stds['ciplv'],ec_stds['imcoh'],ec_stds['wpli2']],
                label='EC')
rects2 = ax.bar(ind + width/2, [eo_means['coh'], eo_means['ciplv'], eo_means['imcoh'],eo_means['wpli2']], width, yerr=[eo_stds['coh'],eo_stds['ciplv'],eo_stds['imcoh'],eo_stds['wpli2']],
                label='EO')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Mean connectivity',fontsize=14,fontweight='bold')
ax.set_xlabel('Methods',fontsize=14,fontweight='bold')
ax.set_xticks(ind)
ax.set_xticklabels(('Coh', 'ciPLV', 'imCoh', 'dwPLI'),fontsize=14)
ax.text(-0.1, .12, f"D={d_vals['coh']:.3f}",fontsize=14)
ax.text(0.8, .03, f"D={d_vals['ciplv']:.3f}",fontsize=14)
ax.text(1.8, 0.01, f"D={d_vals['imcoh']:.3f}",fontsize=14)
ax.text(2.7, 0.01, f"D={d_vals['wpli2']:.3f}",fontsize=14)
ax.legend()
fig.tight_layout()
plt.savefig('figs/real_func_means.png',dpi=300)
```
